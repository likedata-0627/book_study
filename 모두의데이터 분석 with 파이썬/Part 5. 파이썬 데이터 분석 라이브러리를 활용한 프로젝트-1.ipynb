```python
# 조건에 부합하는 데이터는 true 아니면 false
print(a<0)
```

```python
#마스크를 다시 배열에 적용하면 조건에 부합하는 데이터만 남음
print(a[a<0])
```

```python
#마스크를 변수에 담아서 쓸 수도 있음
mask1 = abs(a) >3
print(a[mask1])
```

```python
# 변수에 담은 마스크를 연결해서 사용도 가능
mask1 = abs(a) >3
mask2 = abs(a) %2 ==0
print(a[mask1 + mask2]) #둘중 하나의 조건이라도 참일 경우
print(a[mask1 * mask2]) #두가지 조건이 모두 참일 경우
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.random.randint(-100,100,1000) #1000개의 랜덤값 추출
y= np.random.randint(-100,100,1000) #1000개의 랜덤값 추출
size=np.random.rand(100)*100
mask1 = abs(x)>50 #x에 저장된 값 중 절대값이 50보다 큰 값을 걸러냄
mask2 = abs(y)>50 #y에 저장된 값 중 절대값이 50보다 큰 값을 걸러냄
x=x[mask1+mask2] #mask1과 mask2 중 하나라도 만족하는 값 저장
y=y[mask1+mask2] #mask1과 mask2 중 하나라도 만족하는 값 저장
plt.scatter(x,y,s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
