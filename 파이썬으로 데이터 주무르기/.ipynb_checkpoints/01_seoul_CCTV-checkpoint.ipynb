## 1-6 pandas 고급 두 DataFrame 병합하기

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                  index = [0,1,2,3])

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']},
                  index = [4,5,6,7])

df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']},
                  index = [8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# concat 열방향으로 병합
result = pd.concat([df1,df2,df3])
result
```

```python
# keys 옵션 => 지금은 다중 인덱스가 형성 되어 level을 형성
result = pd.concat([df1,df2,df3], keys=['x','y','z'])
result
```

```python
result.index
```

```python
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

```python
result
```

```python
df4 = pd.DataFrame({
                   'B':['B2','B3','B6','B7'],
                   'D':['D2','D3','D6','D7'],
                   'F':['F2','F3','F6','F7']},
                  index = [2,3,6,7])
# axis=0 => 열기준(기본) / axis=1 => 행기준
result = pd.concat([df1, df4],axis=1 )
```

```python
df1
```

```python
df4
```

```python
result
```

```python
#axis =1 => 행기준을 join=inner => 공통점 만 남겨두세요. 
result = pd.concat([df1,df4],axis=1, join='inner')
result
```

```python
#axis =1 => 행기준을 join_axes = [df1.index] => df1의 인덱스만 만겨두세요. 
result = pd.concat([df1,df4],axis=1, join_axes=[df1.index])
result
```

```python
# ignore_index=True => 두 데이터 프레임의 인덱스를 무시하고 새로운 인덱스를 부여
result = pd.concat([df1,df4], ignore_index=True)
result
```

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# on='key' => 공통된 key값만 합친다. 
pd.merge(left, right, on='key')
```

```python
# 합치는 두데이터를 하나씩 기준으로 합칠때
# how 옵션으로 한쪽 데이터를 설정
pd.merge(left, right, how='left', on='key')
```

```python
pd.merge(left, right, how='right',on='key')
```

```python
# outer => merge한 데이터를 모두 가짐. 
pd.merge(left, right, how='outer',on='key')
```

```python
# innner => 교집합
pd.merge(left, right, how='inner',on='key')
```

## 1-7 cctv 데이터와 인구 데이터 합치고 분석하기

```python
# 열 삭제 => del , 행방향 삭제=> drop
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result.head()
```

```python
data_result.set_index('구별',inplace=True)
data_result.head()
```


* 고령자 비율, 외국인 비율, 인구수 중 무슨 데이터와 cctv를 비교할 지를 정하기
* => 가장 단순한 작업이 상관계수를 조사하는 것
* => 상관계수의 절대값이 클수록 두 데이터는 관계가 있다고 보는 것
* => 0.1 이면 무시, 0.3이하면 약한 상관관계, 0.7 이하면 뚜렷한 상관관계

```python
# corrcoef => 상관계수 계산하기 => 결과가 행렬로 나옴. 
np.corrcoef(data_result['고령자비율'], data_result['소계']) # 무시
```

```python
np.corrcoef(data_result['외국인비율'], data_result['소계']) # 무시
```

```python
# polyfit => 데이터를 대표하는 직선 하나 그리기. 
fp1 =np.polyfit(data_result['인구수'], data_result['소계'],1)
fp1
```

```python
f1 = np.poly1d(fp1) #y축
fx = np.linspace(100000, 700000, 100) #x축
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

## 1-10 조금 더 설득력 있는 자료 만들기

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'],1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

data_result['오차']=np.abs(data_result['소계']-f1(data_result['인구수']))

df_sort = data_result.sort_values(by='오차', ascending =False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['소계'],
           c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,
            df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
