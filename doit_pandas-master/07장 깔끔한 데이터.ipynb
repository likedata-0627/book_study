# 7-2 열이름 관리하기

```python
ebola=pd.read_csv('./datas/data/country_timeseries.csv')
ebola.columns
```

```python
ebola.iloc[:5,[0,1,2,3,10,11]]
```

```python
ebola_long = pd.melt(ebola, id_vars=['Date','Day'])
ebola_long
```

```python
#split으로 열 이름을 분리하기
variable_split= ebola_long.variable.str.split('_')
variable_split[:5]
```

```python
print(type(variable_split))
print(type(variable_split[0]))
```

```python
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)
```

```python
print(status_values[:5])
print(status_values[-5:])
print(country_values[:5])
print(country_values[-5:])
```

```python
#데이터 프레임에 열 추가하기
ebola_long['status'] = status_values
ebola_long['country'] = country_values
ebola_long.head()
```

```python
#concat메서드로 데이터프레임에 추가하기
variable_split = ebola_long.variable.str.split('_', expand=True)
variable_split.columns =['status','country']
ebola_persed = pd.concat([ebola_long, variable_split], axis =1)
ebola_persed.head()
```

# 7-3 여러열을 하나로 정리하기

```python
weather = pd.read_csv('./datas/data/weather.csv')
weather.iloc[:5,:]
```

```python
weather_long =pd.melt(weather, id_vars=['id','year','month','element'], var_name = 'day', value_name='temp')
weather_long
```

```python
#pivot_table 메서드는 행과 열의 위치를 바꿔 정리하는것
weather_tidy = weather_long.pivot_table(
    index=['id','year','month','day'],
    columns = 'element',
    values='temp'
)
weather_tidy
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat
```

# 7-4 중복데이터 처리하기

```python
billboard = pd.read_csv('./datas/data//billboard.csv')
billboard_long = pd.melt(billboard, id_vars=['year','artist','track','time','date.entered'], var_name='week', value_name='rating')
print(billboard_long.shape)
print(billboard_long.head())
```

```python
billboard_long[billboard_long.track=='Loser'].head()
```

```python
# 중복데이터를 가지고 있는 열만 따로 모아서 변수에 담는다
billboard_songs = billboard_long[['year','artist','track','time']]
billboard_songs
```

```python
#drop_duplicates 메서드로 중복된 데이터 제거
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

```python
billboard_songs['id']=range(len(billboard_songs))
billboard_songs.head(n=10)
```

```python
#노래 정보와 주간 순위 데이터를 합친것.
billboard_ratings = billboard_long.merge(billboard_songs, on=['year','artist','track','time'])
print(billboard_ratings.shape)
print(billboard_ratings.head())
```
