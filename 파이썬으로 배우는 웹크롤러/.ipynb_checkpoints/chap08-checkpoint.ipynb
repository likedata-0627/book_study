### 돔 접근 함수
>* selenium으로 돔 접근을 할 경우 bs4와는 차이가 있음
>* 반환 타입: 단일 객체 (bs4에서 find() 같은 형태로 반환)
>>* find_element_by_id
>>* find_element_by_tag_name
>>* find_element_by_xpath //xml문서에 접근하기 위한 문법
>>* find_element_by_link_text
>>* find_element_by_partial_link_text
>>* find_element_by_name
>>* find_element_by_class_name
>>* find_element_by_css_selector
>* 반환 타입: 리스트(bs4에서 find_all() 같은 형태로 반환)
>>* find_elements_by_id
>>* find_elements_by_tag_name
>>* find_elements_by_xpath //xml문서에 접근하기 위한 문법
>>* find_elements_by_link_text
>>* find_elements_by_partial_link_text
>>* find_elements_by_name
>>* find_elements_by_class_name
>>* find_elements_by_css_selector

### 8-4 태그 이용하여 접근
>* 태그 이름을 이용해돔 접근이 가능
>* 태그 이름을 이용해 돔을 가져올 때는 하나의 돔 또는 여러개의 돔을 리스트 형태로 가져올 수 있음
>* elements로 리스트를 반환 할때는 돔의 세션을 리스트로 만들어 반환해줌. bs4 에서 find()와 find_all()과 같은 형식으로 반환
>* find_elements_by_tag_name() 함수는 리스트이므오 text와 tag_name 속성을 사용할 수 없음
>* 각각 요소의 tag_name 또는 text 속성에 접근하고 싶으면 반복문을 사용하면 됨

```python
from selenium import webdriver

url = 'https://pjt3591oo.github.io'
driver = webdriver.Chrome('chromedriver') 
driver.get(url) 

selected_id = driver.find_element_by_id('nav-trigger')
print(selected_id)
print(selected_id.tag_name)
print(selected_id.text)

seleceted_tags_p = driver.find_elements_by_tag_name('p')
print(seleceted_tags_p)
```

### 8-13 click 이용하여 페이지 이동
>* 에러가 발생하는 이유는 메인페이지에서 돔을 가져오고 다른 페이지로 넘어가면 메인 페이지에서 가져온 돔에 접근 할 수없기 때문임
>* 고로 click을 이용하여 페이지를 이동하는 것을 비추
>* 클릭을 사용하는 경우 클릭했을 때, 페이지 변화가 없고 해당 페이지 내에서 데이터가 추가되는 경우에 사용하는 것이 좋음

```python
from selenium import webdriver

url = 'https://pjt3591oo.github.io'

driver = webdriver.Chrome('chromedriver')
driver.get(url)

selected_tags_a = driver.find_elements_by_css_selector('a')

for i in selected_tags_a:
    print(i.text, i.tag_name)
    i.click()
```

### 8-19 selenium vs BeautifulSoup 
>* selenium은 bs4와 같이 사용해야 최고의 효율을 낼 수 있음
>* selenium이 bs4보다 더 오래 걸림 => 이유는 웹 브라우저를 띄우는데 많은 자원을 사용하기 때문
>* 그러므로 selenium으로 requests로 처리하기 어려운 javascript를 처리 하고, 만들어진 html은 bs4를 이용하여 처리

```python
# selenium 사용 시간 측정
from selenium import webdriver
import time

url = 'https://pjt3591oo.github.io'

selenium_start = time.time()

driver = webdriver.Chrome('chromedriver')
driver.get(url)

div_selectors11 = driver.find_element_by_css_selector('div')
div_selectors12 = driver.find_element_by_css_selector('div')
div_selectors13 = driver.find_element_by_css_selector('div')
div_selectors14 = driver.find_element_by_css_selector('div')
div_selectors15 = driver.find_element_by_css_selector('div')

selenium_end = time.time() -selenium_start

print('usage selenium :%f' %(selenium_end))

# bs4 사용시간 측정
import requests as rq
from bs4 import BeautifulSoup

bs4_start = time.time()

res = rq.get(url)
soup = BeautifulSoup(res.content, 'lxml')
div_selectors21 = soup.select('div')
div_selectors22= soup.select('div')
div_selectors23 = soup.select('div')
div_selectors24 = soup.select('div')
div_selectors25 = soup.select('div')

bs4_end = time.time() -bs4_start

print('usage bs4 : %f' %(bs4_end))
```

### 8-23 검색 결과 데이터 수집
>* javascript를 이용해서 html을 추가하는 대표적인 예가 검색임
>* 이럴 때는 개발자 도구[network]탭에서 서버에 검색하는 부분을 찾아 직접 요청 할수 있지만, selenium 같은 경우에는 굳이 개발자 도구를 이용안해도 됨.

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

from bs4 import BeautifulSoup

url = 'https://pjt3591oo.github.io/search'

search_keysword = 'db'

driver = webdriver.Chrome('chromedriver')
driver.get(url)

selected_tag_a = driver.find_element_by_css_selector('input#search-box')

selected_tag_a.send_keys(search_keysword)
selected_tag_a.send_keys(Keys.ENTER)  # ‘\ue007’로 해도 엔터가 됨

soup = BeautifulSoup(driver.page_source, 'lxml')
items = soup.select('u#search-result li')

for item in items:
    title = item.find('h3').text
    description = item.find('p').text
    print(title)
    print(description)
```
